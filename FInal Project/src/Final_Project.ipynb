Data Preprocessing, Create A Sharded TFRecord Files

```python
import tensorflow as tf
import os
from glob import glob
from PIL import Image
import numpy as np
import random

def load_image(image_path):
    """Load an image as a NumPy array."""
    image = Image.open(image_path)
    return np.array(image)

def create_sharded_tfrecord(data_dir, output_prefix, num_samples, num_shards):
    """
    Creates sharded TFRecord files from a dataset with a fixed number of samples.

    Args:
        data_dir (str): Directory containing the dataset folders.
        output_prefix (str): Prefix for the output TFRecord shards.
        num_samples (int): Number of samples to select from the dataset.
        num_shards (int): Number of shards to create.
    """
    folders = glob(os.path.join(data_dir, "*"))
    all_records = []

    # Gather all valid image triplets
    for folder in folders:
        images = sorted(glob(os.path.join(folder, "*.png")))
        if len(images) < 3:
            continue
        all_records.append((images[0], images[1], images[2]))

    # Randomly sample the desired number of records
    if len(all_records) > num_samples:
        all_records = random.sample(all_records, num_samples)

    # Divide records into shards
    records_per_shard = len(all_records) // num_shards
    for shard_idx in range(num_shards):
        shard_filename = f"{output_prefix}-{shard_idx:05d}-of-{num_shards:05d}"
        writer = tf.io.TFRecordWriter(shard_filename)

        # Define start and end indices for this shard
        start_idx = shard_idx * records_per_shard
        end_idx = len(all_records) if shard_idx == num_shards - 1 else (shard_idx + 1) * records_per_shard

        for record in all_records[start_idx:end_idx]:
            frame1, frame2, frame3 = load_image(record[0]), load_image(record[1]), load_image(record[2])
            feature = {
                "frame1": tf.train.Feature(bytes_list=tf.train.BytesList(value=[frame1.tobytes()])),
                "frame2": tf.train.Feature(bytes_list=tf.train.BytesList(value=[frame2.tobytes()])),
                "frame3": tf.train.Feature(bytes_list=tf.train.BytesList(value=[frame3.tobytes()])),
                "height": tf.train.Feature(int64_list=tf.train.Int64List(value=[frame1.shape[0]])),
                "width": tf.train.Feature(int64_list=tf.train.Int64List(value=[frame1.shape[1]])),
            }
            example = tf.train.Example(features=tf.train.Features(feature=feature))
            writer.write(example.SerializeToString())
        writer.close()

    print(f"Sharded TFRecord files written to {output_prefix}-*")

# Usage
create_sharded_tfrecord("/content/ATD12K/datasets/train_10k", "train_subset", num_samples=1, num_shards=1)
create_sharded_tfrecord("/content/ATD12K/datasets/test_2k_540p", "test_subset", num_samples=1, num_shards=1)
```
